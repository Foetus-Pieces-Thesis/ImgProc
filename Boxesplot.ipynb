```python
from skimage.measure import regionprops, label
import skimage.io
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle as drawrect
import numpy as np

%matplotlib qt

z =5   # int(input("how many z slices do you have in your movie?"))
t = 64 #int(input("how many time slices do you have in your movie?"))
masks_list = []
labels = []
Regions = []

for j in range(t):
    Regions_temp = []
    masks_list_temp = []
    labels_temp = []
    for i in range(1, z+1):
        mask = skimage.io.imread("C1-movie-%i_cp_masks.tif" % (int(i)))
        masks_list_temp.append(mask[j])
        labelled = label(mask[j])
        labels_temp.append(labelled)
        Regions_temp.append(regionprops(labelled))
    Regions.append(Regions_temp)
    masks_list.append(masks_list_temp)
    labels.append(labels_temp)




print(len(Regions[0][1]))


# Regions = np.reshape(Regions, (64, 5))
# masks_list = np.reshape(masks_list, (64, 5))
# labels = np.reshape(labels, (64, 5))


data = skimage.io.imread("movie.tif")

# this just requires you to load the movie

data = np.moveaxis(data, 3, 0)
data = np.moveaxis(data, 4, 1)

cells = data[:,:,:,:,0]

Areas = []
BBoxes = []
Centroids = []

for i in range(t):
    Areas_temp = []
    BBoxes_temp = []
    Centroids_temp = []
    for j in range(z):
        Areas_temp.append([thisRegion.area for thisRegion in Regions[i][j]])
        BBoxes_temp.append([thisRegion.bbox for thisRegion in Regions[i][j]])
        Centroids_temp.append([thisRegion.centroid for thisRegion in Regions[i][j]])
    Areas.append(Areas_temp)
    BBoxes.append(BBoxes_temp)
    Centroids.append(Centroids_temp)


np.random.seed(19680801)

class IndexTracker:
    def __init__(self, ax, X, areas, bboxes, centroids):
        self.ax = ax
        ax.set_title("use arrow keys. up/down = scroll z-stack, left/right = scroll time series")

        self.X = X
        self.areas = areas
        self.bboxes = bboxes
        self.centroids = centroids
        rows, cols, self.tslices, self.zslices = X.shape
        self.zind = self.zslices // 2
        self.tind = self.tslices // 2

        # requires [x, y, t, z] shape
        self.im = ax.imshow(self.X[:, :, self.tind, self.zind])
        self.update()

    def on_key_press(self, event):
        print("%s" % (event.key))
        if event.key == 'up':
            self.zind = (self.zind + 1) % self.zslices
        if event.key == 'down':
            self.zind = (self.zind - 1) % self.zslices
        if event.key == 'left':
            self.tind = (self.tind - 1) % self.tslices
        if event.key == 'right':
            self.tind = (self.tind + 1) % self.tslices
        # add try except here to account for not necessarily having all 3 channels
        self.ax.clear()
        self.update()

    def update(self):
        ax.set_title("use arrow keys. up/down = scroll z-stack, left/right = scroll time series")
        self.im.set_data(self.X[:, :, self.tind, self.zind])
        self.im = ax.imshow(self.X[:, :, self.tind, self.zind])
        self.ax.set_ylabel('z-slice %s, t-slice %s' % (self.zind, self.tind))
        for (bb, area, centroid) in zip(self.bboxes[self.tind][self.zind], self.areas[self.tind][self.zind], self.centroids[self.tind][self.zind]):
            r = drawrect((float(bb[1] - 1), float(bb[0] - 1)), bb[3] - bb[1], bb[2] - bb[0], edgecolor='r', facecolor='none')
            self.ax.add_patch(r)
            plt.scatter(centroid[1] - 1, centroid[0] - 1, c='r', s=2)
        self.im.axes.figure.canvas.draw()

fig, ax = plt.subplots(1, 1)



tracker = IndexTracker(ax, cells, Areas, BBoxes, Centroids)

fig.canvas.mpl_connect('key_press_event', tracker.on_key_press)
plt.show()


# Areas = [thisRegion.area for thisRegion in Regions]
# BBs = [thisRegion.bbox for thisRegion in Regions]
# Centroids = [thisRegion.centroid for thisRegion in Regions]
# fig, ax = plt.subplots()
# ax.imshow(cells)
# ax.axis('off')
# for (bb,area,centroid) in zip(BBs,Areas,Centroids):
#     r = drawrect((float(bb[1]-1),float(bb[0]-1)), bb[3]-bb[1],bb[2]-bb[0], edgecolor='r', facecolor='none')
#     ax.add_patch(r)
#     plt.scatter(centroid[1]-1, centroid[0]-1, c='r', s=2)
```
